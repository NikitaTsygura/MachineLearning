```python
import numpy as np
import pandas as pd

coins_input_str = input("Введите количество монет за 5 дней, разделенных запятыми (например, 10,12,8,15,20): ")
coins_list = [int(coin.strip()) for coin in coins_input_str.split(',')]

data = {
    "Days": [1, 2, 3, 4, 5],
    #"Coins": [2, 2, 6, 15, 10]
    "Coins":coins_list
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Days"], df["Coins"], color="yellow")

plt.title("График залежности")
plt.xlabel("Days")
plt.ylabel("Coins")

plt.show()
```

**Вчимо та даэмо формулу**

```python
# scikit-learn
from sklearn.linear_model import LinearRegression

X =df[["Days"]]
Y =df[["Coins"]]

model = LinearRegression()
model.fit(X, Y)

print("Bara (W):", model.coef_)
print("Змищення (b):", model.intercept_)
```

```python
Y_pred = model.predict(X)

plt.scatter(X, Y, color="red")
plt.plot(X, Y_pred, color="green")
plt.title("Линийна регресия")
plt.xlabel("Days")
plt.ylabel("Coins")
```

**Прогноз**

```python
#new_value = [[10]]

newvalue_input_str = input("Введите день на прогноз: ")
newvalue_list = [int(day) for day in newvalue_input_str.split(',')]

predicted_coins = model.predict(newvalue_list)
print(f"Прогноз на {newvalue_list[0][0]}: {predicted_coins[0]} монеток")
```
